## Final Heatmap

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns

with open('github_repos_with_users.json') as json_file:
    data = json.load(json_file)
```

```python
df = pd.json_normalize(data, record_path='mentionableUsers', meta=['name', 'updatedAt', 'url'], record_prefix='user_')
```

```python
df = df[df.user_login != 'dominikb1888']
df = df[df.user_login != 'ProfEnergyHuber']
```

```python
# df.apply(lambda x: x['Formula'].replace('Length', str(x['Length'])), axis=1)

# Remove String 'user_login' from 'name'
df['new'] = df.apply(lambda x: x['name'].replace((x['user_login']), ''), 1)
df['session_no'] = df['new'].str.split('-', expand=True)[0]
df['exercise_no'] = df['new'].str.split('-', expand=True)[1]
df['exercise_name'] =  df['new'].str.split('-', expand=True)[2] + ' ' +  df['new'].str.split('-', expand=True)[3]
```

```python
dff = df[['session_no', 'exercise_no', 'exercise_name', 'user_login', 'url','updatedAt']]
dff = (dff
  .astype({
     'session_no':  'int8',
     'exercise_no': 'int8',
     'exercise_name': 'category',
     'user_login': 'category',
     'updatedAt': 'datetime64[ns]',
  })
  .sort_values('updatedAt', ascending=True)
)
dff
```

```python
print("Memory df:", df.memory_usage(deep=True).sum())
print("Memory dff:", dff.memory_usage(deep=True).sum())
```

```python
dff
```

```python
pivot_table = dff.pivot_table(index="user_login", columns="session_no", values='url', aggfunc='count')
pivot_table[pivot_table == 0] = np.nan
pivot_table['sum_cols'] = pivot_table.sum(axis=1)
pivot_table = pivot_table.sort_values('sum_cols', ascending=False).iloc[0:20]
pivot_table.drop('sum_cols', axis=1, inplace=True)
pivot_table
```

```python
sns.set(rc={'figure.figsize':(11,44)})

heatmap_plot = sns.heatmap(
    pivot_table,
    annot_kws = {
        'fontsize': 11,
        'va':'center_baseline',
    },
    vmin = 1,
    vmax = 17,
    cmap = 'crest',
    annot = True,
    linecolor = 'white',
    linewidths = .6,
    cbar = False,
    square = True,
)
heatmap_plot.set_xlabel('')
heatmap_plot.set_ylabel('')
heatmap_plot.set_yticklabels(heatmap_plot.get_yticklabels(), va='center_baseline')
heatmap_plot.set_xticklabels(heatmap_plot.get_xticklabels(), ha='center')
heatmap_plot.xaxis.tick_top()


```

```python
fig = heatmap_plot.get_figure()
fig.savefig("heatmap.png") 
```

## Animated Heatmap

```python
sns.set(rc={'figure.figsize':(11,44)})
for i in range(1, 300):
    plt.clf()
    
    start_date = datetime.strptime("2022-03-16 08:00:00", '%Y-%m-%d %H:%M:%S')
    pivot_table = (dff[dff.updatedAt < start_date + timedelta(days = i)]
                   .pivot_table(index="user_login", columns="session_no", values='url', aggfunc='count'))
    pivot_table[pivot_table == 0] = np.nan
    pivot_table['sum_cols'] = pivot_table.sum(axis=1)
    pivot_table = pivot_table.sort_values('sum_cols', ascending=False).iloc[0:20]
    pivot_table.drop('sum_cols', axis=1, inplace=True)

    heatmap_plot = sns.heatmap(
        pivot_table,
        annot_kws = {
            'fontsize': 11,
            'va':'center_baseline',
        },
        vmin = 1,
        vmax = 17,
        cmap = 'crest',
        annot = True,
        linecolor = 'white',
        linewidths = .6,
        cbar = False,
        square = True,
        )
    
    heatmap_plot.set_xlabel('')
    heatmap_plot.set_ylabel('')
    heatmap_plot.set_yticklabels(heatmap_plot.get_yticklabels(), va='center_baseline')
    heatmap_plot.set_xticklabels(heatmap_plot.get_xticklabels(), ha='center')
    heatmap_plot.xaxis.tick_top()
    
    fig = heatmap_plot.get_figure()
    fig.savefig(f"heatmaps/{i}_heatmap.png") 

    # then convert on commandline to animated gif...
    #
    # $ convert -delay 35 -loop 0 *_heatmap.png heatmap_animated.gif
    # $ mogrify -layers 'optimize' -fuzz 7% heatmap_animated.gif
    
#anim = animation.FuncAnimation(fig, animate, init_func=init, frames=frames, interval=1, repeat=True)
#f = r"heatmap_animation.gif" 
#writergif = animation.PillowWriter(fps=30) 
#anim.save(f, writer=writergif)
```

## Barchart

```python
sns.set(rc={'figure.figsize':(11,5)})
barchart_data = dff.pivot_table(values="url", index=["session_no","exercise_no"], aggfunc='count').reset_index()
barchart_data['id'] = barchart_data['session_no'].astype(str) + '_' + barchart_data['exercise_no'].astype(str)
barchart_data = barchart_data[barchart_data.url > 4]
```

```python
bar = sns.barplot(
    data=barchart_data, 
    x='id', 
    y="url", 
    orient='v',
    )
plt.xticks(rotation=90, fontsize=8)
bar
```

## Timeline Heatmap

```python
timeline_data = dff.pivot_table(
    index='user_login', 
    columns=pd.Grouper(freq='W',key='updatedAt'), 
    values='exercise_no', 
    aggfunc='count',
)
```

```python
timeline_data['total'] = timeline_data.sum(axis=1)
timeline_data = timeline_data.sort_values('total' , ascending=False)
timeline_data.drop('total', axis=1, inplace=True)
timeline_data
```

```python
sns.set(rc={'figure.figsize':(11,44)})
sns.heatmap(
    data=timeline_data, 
    annot_kws = {
        'fontsize': 9,
        'va':'center_baseline',
    },
    vmin = 1,
    vmax = 14,
    cmap = 'crest',
    linecolor = 'white',
    linewidths = .6,
    annot = True,
    cbar = False,
    square = True,
)
```
